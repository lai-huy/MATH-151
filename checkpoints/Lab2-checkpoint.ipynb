```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

# 1a volume of ice cream in cone

```python
h = 4
theta_rad = 30 * pi/180
r = h * sin(theta_rad/2)
vol_cone = 1/3 * pi * r**2 * h
vol_hemi = 2/3 * pi * r ** 3
vol = vol_cone + vol_hemi
vol.evalf()
```

# 1b slant height needed for one cup

```python
vol = 14.5
h = symbols('h')
r = h * sin(theta_rad/2)
vol_cone = 1/3 * pi * r**2 * h
vol_hemi = 2/3 * pi * r ** 3
solve(vol_cone + vol_hemi - vol, h)
```

# 3b intersections of circles

```python
s = symbols('s')
t = symbols('t')
x1 = 0 + 7 * cos(t)
y1 = 0 + 7 * sin(t)
x2 = 3 + 5 * cos(s)
y2 = 4 + 5 * sin(s)
eq_x = x1 - x2
eq_y = y1 - y2
sol = solve((eq_x, eq_y), (s, t))
s0 = sol[0][0]
s1 = sol[0][1]
t0 = sol[1][0]
t1 = sol[1][1]

# Verified using Desmos
print("circle 1: x=", x1.subs(t, t1).evalf())
print("circle 1: y=", y1.subs(t, t1).evalf())
print("circle 2: x=", x2.subs(s, s0).evalf())
print("circle 2: y=", y2.subs(s, s0).evalf())
```
