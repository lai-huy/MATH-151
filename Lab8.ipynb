# 1a find and simplify the derivative of f-g

```python
x = symbols('x', positive=True)
f = 2 * asin(x)
g = acos(1 - 2 * x ** 2)
h = (f - g).diff(x)
print("The derivative is", h)
print("Simplified, it becomes,", simplify(h), "when x is non-negative")
```

# 1bc Implication of f-g and specific answer

```python
print("When x is non-negative, f(x) - g(x) is 0, implying that f(x) and g(x) are equivalent.")
print("f(0) - g(0) =", f.subs(x, 0) - g.subs(x, 0))
print("The two functions are equivalent")
```

# 3a Plot of f and number of extrema and inflection points based on graph

```python
matplotlib notebook
```

```python
x = symbols('x', real=True)
f = (x ** 2 * (x + 1) ** 3)/((x - 2) ** 2 * (x - 4) ** 4)
plot(f, xlim=[-10,10])
print("There appears to be 3 local extrema and 4 inflection points")
```

# 3b f' and critical values

```python
dfdx = f.diff(x)
dfdx = simplify(dfdx)
sols = solve(dfdx)
print("Derivative: ", dfdx)
print("Critical Values:")
for sol in sols:
    print(re(sol.evalf()))
```

# 4a Define f and g for ln(y)
