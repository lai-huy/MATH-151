```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

# 1a Tangent line at x=2

```python
x = symbols('x')
f = 5 * x / (1 + x**2)
m = f.diff().subs(x, 2)
print(m)
```

# 1b Graph of f and tangent line

```python
matplotlib notebook
```

```python
tan_line = m * (x - 2) + f.subs(x, 2)
plot(f, tan_line, xlim=[-4,4])
```

# 2a Veclocity after 1 second

```python
t = symbols('t')
s = 1 + 10 * t - 1.86 * t ** 2
v = s.diff()
v.subs(t, 1)
```

# 2b when v=0 and corresponding height

```python
solve(v, t)
```

# 2c velocity when hits the surface

```python
v.subs(t, solve(s, t)[1])
```

# 3b Equations of horizontal tangent lines

```python
f_pos = x * exp(-x)
f_neg = x * exp(x)
print(solve(f_pos.diff(), x), solve(f_neg.diff(), x))
```

# 4a first 16 derivatives

```python
x = symbols('x')
f = x * sin(x)
diffs = [f.diff(x,n) for n in range(1,17)]
print(diffs)
```
