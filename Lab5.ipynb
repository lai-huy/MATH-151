# 1a Values of r which solve differential equation

```python
x = symbols('x')
r = symbols('r')

y = exp(r * x)
dy = y.diff(x)
ddy = dy.diff(x)
solve(ddy + 5 * dy + 6 * y, r)
```

```python
y = exp(r * x)
dy = y.diff(x)
ddy = dy.diff(x)
solve(ddy + 6 * dy + 13 * y, r)
```

# 1c Compute left hand side with cosines and sines and conclusion

```python
y = exp(-3 * x) * (cos(2 * x) + sin(2 * x))
dy = y.diff(x)
ddy = dy.diff(x)
simplify(ddy + 6 * dy + 13 * y)

# e^(-3x)(cos(2x)+sin(2x)) is a solution to the differential equation
```

# 2a Implicit Plot

```python
matplotlib notebook
```

```python
x = symbols('x', real=True)
y = symbols('y', real=True)
F = (x**2 + y**2)**2 - (4 * (x**2 - y**2)) + 3
Fplot = plot_implicit(F, (x, -3, 3),(y,-3,3))
```

# 2b dy/dx

```python
dydx = idiff(F, y, x)
print("The derivative of the function is", dydx)
```

# 2c Horizontal tangent lines

```python
num = numer(dydx)
htan = solve([num, F],[x,y])
print("The horizontal tangents are", htan)
```

# 2d Tangent line equation at (sqrt(2), sqrt(sqrt(17)-4))

```python
point = {x:sqrt(2), y:sqrt(sqrt(17) - 4)}
m = dydx.subs(point)
line = m * (x - point[x]) + point[y]
print("The tangent line equation is", line)
```

# 2e Plot equation with tangent line in part (d)

```python
matplotlib notebook
```

```python
pcurve = plot_implicit(F, (x, -3, 3), (y, -3, 3),show=False)
pcurve.extend(plot(line, show=False))
pcurve.show()
```
