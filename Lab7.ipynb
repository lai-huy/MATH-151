# 2a Critical values of f

```python
x = symbols('x', real=True)
f = exp(x) + exp(-3 * x)
cvals = solve(f.diff(), x)
print(cvals)
print([i.evalf() for i in cvals])
```

# 2b absolute max and min on [0,1]

```python
candidates = [0, 1]
for val in cvals:
    if 0 <= val <= 1:
        candidates.append(val)

vals = []
for val in candidates:
    vals.append(f.subs(x, val))
print(max(vals).evalf())
print(min(vals).evalf())
```

# 4a Two-variable function to be maximized and equation relating the two variables

```python
s = symbols('s')
h= symbols('h')
SA = symbols('SA')
A = (s ** 2) + 4 * (h * s)
V = s * s * h
print(V)
print(A)
A = (s ** 2) + 4 * (h * s) - SA
```

# 4b Solve equation and substitute into function

```python
height = solve(A, h)
V = V.subs(h, height[0])
print(V)
```

# 4c practical domain

```python
domains = solve(V, s)
print("The upper bound of the domain is",domains[2])
```

# 4d Absolute maximum of function from part (b) on domain from part (c)

```python
maxium = solve(V.diff(s), s)
print('The maxium side length is', V.subs(s, maxium[1]))
A = A.subs(s, maxium[1])
max_Height = solve(A, h)
print('The maxium height is', max_Height[0])
```
