# 2a Plot V(d)

```python
matplotlib notebook
```

```python
d = symbols('d')
M = 0.1
p_w = 1000.0
C = 1.0
theta = 10.0
beta = 10.0
g = 9.81
theta *= pi/180
beta *= pi/180
V = (sqrt((16 * M * g)/(pi * C * p_w * d**2)))/(sqrt(1 - (8 * M * tan(beta)**2)/(pi * d**3 * C * p_w * sin(theta))))
plot(V, xlim = [.05, .1], ylim = [.7, .9])
```

# 2b Theorem to find solution to V(d)=0.8

```python
a = .09
b = .095

print("Using the Intermediate Value Theorm")
print("a =", a)
print("b =", b)
```

# 2c Numerical Solution

```python
print(nsolve(V-.8, a))
```

# 3a List Comprehension

```python
x = symbols('x')
f = sqrt(2 * x**2 + 1) / (3 * x - 5)
print([f.subs(x, float(i)) for i in [10, 50, 100]])
print([f.subs(x, float(i)) for i in [-10, -50, -100]])
    
```

# 3b Compute Limits

```python
pos_inf = limit(f, x, oo)
neg_inf = limit(f, x, -oo)

print(neg_inf)
print(pos_inf)
print(float(neg_inf))
print(float(pos_inf))
```

# 3c Plot function and Horizontal Asymptotes

```python
matplotlib notebook
```

```python
plot(f, pos_inf, neg_inf, xlim = [-10,10], ylim = [-5,5])
```

# 4a Graph of piecewise function

```python
matplotlib notebook
```

```python
x=symbols('x')
fa=2*x-x**2
fb=2-x
fc=(x**2 -7*x+12)/(x-3)
fd= pi
plot((fa,(x, 0,2)),(fb,(x, 2,3)),(fc,(x, 3,4)),(fd,(x, 4,6)))
# f is continuous at 2 and 3 but not 4
```

# 4b Left and Right Hand Limits at 2, 3, and 4

```python
lim_2l = limit(fa,x,2)
lim_2r = limit(fb,x,2)
print("Left hand limit at 2: ", lim_2l)
print("Right hand limit at 2: ",lim_2r)
lim_3l = limit(fb,x,3)
lim_3r = limit(fc,x,3)
print("Left hand limit at 3: ", lim_3l)
print("Right hand limit at 3: ",lim_3r)
lim_4l = limit(fc,x,4)
lim_4r = limit(fd,x,4)
print("Left hand limit at 4: ", lim_4l)
print("Right hand limit at 4: ", lim_4r)
```
